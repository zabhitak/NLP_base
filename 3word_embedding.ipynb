Word Embedding Techniques using Embedding Layer in Keras

```python
### Libraries USed Tensorflow> 2.0  and keras
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#sentences
#for better embidding all sentences should have same words
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
#use pad_sequences to make same size
```

```python
#vocabluary size
voc_size = 10000
```

One hot Representation

```python
onehot_rep = [one_hot(words,voc_size)for words in sent]
onehot_rep
#all this number will be < 10000
```

Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
 sent_length = 8
 embedded_docs = pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
 embedded_docs
```

```python
#now choose features 
dimensions = 10
model= Sequential()
model.add(Embedding(voc_size,dimensions,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```

```python
print(embedded_docs[0])
print(model.predict(embedded_docs[0]))
```
